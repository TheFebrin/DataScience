```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
from src.datareader import DataReader
from src.fastspd import FastSPD
```

```python
def save_to_file(data: np.ndarray, dir = "./VED_dataset/preprocessed.npy"):
    np.save(dir, data, allow_pickle=True)

def load_from_file(dir: str = "./VED_dataset/preprocessed.npy") -> np.ndarray:
    np.load(dir, allow_pickle=True)

def plot_trajectory(trajectory: np.ndarray):
    X = trajectory[:, 1]
    Y = trajectory[:, 2]
    plt.figure(figsize=(10, 5))
    plt.scatter(X, Y)
    plt.title("Trajectory with {} points".format(len(X)))
    plt.show()
```

## Reading the raw data

```python
data_reader = DataReader("./VED_dataset/*.csv", files_limit=100)
data = data_reader.get_data()
trajectory = data[250]
```

```python
plot_trajectory(trajectory)
```

## Testing FastSPD on a single trajectory

```python
fastspd = FastSPD()
compressed_trajectory = fastspd.compress(trajectory)
```

```python
plot_trajectory(compressed_trajectory)
```

## Running FastSPD and saving trajectories to a file

```python
fastspd.compress_and_save(data)
```

## Loading the compressed data from a file

```python
compressed_data = fastspd.load_compressed()
```

```python
plot_trajectory(compressed_data[250])
```
