#### a) 
**Zrób rysunek przedstawiający 10 punktów o następujących współrzędnych (1, 10), (2, 10),
(3, 11), (4, 12), (5, 18), (6, 18), (7, 19), (8, 26), (9, 19), (10, 26).**


```python
points = np.array([(1, 10), (2, 10), (3, 11), (4, 12), (5, 18), (6, 18), (7, 19), (8, 26), (9, 19), (10, 26)])
X, Y = points[:, 0], points[:, 1]
P = plt.figure().gca()
P.scatter(X, Y)
```

#### b) 
**Dodaj do rysunku prostą o równaniu y = 2 x + 5.**

```python
P.plot(X, 2 * X + 5, c='red')
P.figure
```

#### d) 
**Dla każdego punktu danych policz jego odległość od wyznaczonej prostej.**

```python
def dist(X, Y, a, b):
    A, B, C = -a, 1, -b
    P = np.array([X,Y])
    v = np.array([A, B]).reshape(2,1)
    ans = np.abs(np.sum(P * v, axis=0) + C) / np.sqrt(A ** 2 + B ** 2)
    
    return ans


a, b = 2, 5

distances = dist(X, Y, a, b)
d = np.sum(distances)
print(distances)
print(f'Sum of distances: {d}')
```

#### e) 
**Spróbuj zmienić równanie prostej z punktu b) tak, aby zmniejszyć sumę odległości punktów
danych od prostej.**

```python
better_a, better_b = 0, 0
for a in np.arange(-10.0, 10.0, 0.1):
    for b in np.arange(-10.0, 10, 0.1):
        new_distances = dist(X, Y, a, b)
        new_dist = np.sum(new_distances)
        if new_dist < d:
            d = new_dist
            distances = new_distances
            better_a, better_b = a, b

print(distances)
print(f'Best of distances: {d}')
```

```python
P.plot(X, X * better_a + better_b, c='green')
P.figure
```
