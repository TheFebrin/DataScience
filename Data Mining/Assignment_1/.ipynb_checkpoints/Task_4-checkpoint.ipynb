```python
# Set up
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.lines as lines

from matplotlib.ticker import MaxNLocator
from matplotlib.colors import ListedColormap

from sklearn import datasets
iris = datasets.load_iris() 
```

#### a) 
**Zrób rysunek podobny do tego z poprzedniego zadania, ale umieść na nim tylko irysy gatunku
setosa i versicolor (nie rysuj irysów gatunku virginica).**


```python
sepal_length = iris.data[:,0][:100]
sepal_width = iris.data[:,1][:100]

cmap_bold = ListedColormap(['#FF0000', '#0000FF', '#00FF00'])

iris_plot = plt.figure().gca()
iris_plot.set_xlim(4, 7.5)
iris_plot.set_ylim(1.5, 5)

iris_y = np.array(iris['target'][:100])

iris_plot.scatter(sepal_length, sepal_width, c=iris_y, s=15, cmap=cmap_bold)
```

#### b) 
**Dodaj do rysunku prostą o równaniu y = 2 x – 8.**

```python
l = np.linspace(np.min(sepal_length), np.max(sepal_length))
iris_plot.plot(l, l * 2 - 8, c='blue')

iris_plot.figure
```

#### c) 
**Irysy gatunku setosa znajdujące się pod narysowaną linią zaznacz na czerwono, pozostałe na
zielono. Irysy gatunku virginica znajdujące się nad narysowaną linią zaznacz na czerwono,
pozostałe na zielono.**


```python
iris_plot = plt.figure().gca()
iris_plot.set_xlim(4, 7.5)
iris_plot.set_ylim(1.5, 5)
l = np.linspace(np.min(sepal_length), np.max(sepal_length))
iris_plot.plot(l, l * 2 - 8, c='blue')

X = sepal_length
Y = sepal_width

iris_y = np.array(iris['target'][:100])

iris_y = [iris_y[i] ^ 1 if 2 * X[i] - 8 < Y[i] 
          else iris_y[i] for i in range(len(iris_y))]

iris_plot.scatter(X, Y, c=iris_y, s=15, cmap=cmap_bold)
```

#### d)
**Zapisz rysunek do pliku zadanie2a.png.**

```python
iris_plot.figure.savefig('Task2a.png')
```

### e) 
**Spróbuj zmienić równanie prostej z punktu b) tak, aby zmniejszyć liczbę czerwonych punktów.**

```python
z = np.array([1.4, -4.33])
fit_poly = np.poly1d(z)
fit = fit_poly(X)

iris_plot.plot(X, fit, c='pink')
iris_plot.figure
```
