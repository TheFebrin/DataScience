```python
# set up
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
import matplotlib.transforms as transforms
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import proj3d
import random

from sklearn import datasets
import pandas as pd
from sklearn.decomposition import PCA
```

```python
mu2 = [[-10, -10, -4], [0, 0, 0], [2, 4, 20]]
sigma = [[3, 3, 2], 
         [3, 14, 4],
         [2, 4, 5]]

X = np.array([np.random.multivariate_normal(mu2[random.randint(0, 2)], sigma) \
               for _ in range(1000)])

print(X.shape)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(X[:, 0], X[:, 1], X[:, 2], 'o', markersize=8, color='blue', alpha=0.5)
```

```python
X_std = (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(X[:, 0], X[:, 1], X[:, 2], '^', markersize=8, color='blue', alpha=0.5)
ax.plot(X_std[:, 0], X_std[:, 1], X_std[:, 2], 'o', markersize=8, color='red', alpha=0.5)
```

```python
Sigma = np.cov(X_std.T)
v, w = LA.eig(Sigma)
X_pca = np.dot(X_std, np.dot(w, np.diag(1 / np.sqrt(v))))
```

```python
X_pca = w.T.dot(X_std.T).T / np.sqrt(v)

fig = plt.figure(figsize=(20,8))
ax = fig.add_subplot(121, projection='3d')
ax.plot(X_std[:, 0], X_std[:, 1], X_std[:, 2], 'o', markersize=8, color='red', alpha=0.5)
ax.plot(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], '^', markersize=8, color='blue', alpha=0.5)

pca = PCA(n_components=3, whiten=True)
lib = pca.fit_transform(X_std)
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot(X_std[:, 0], X_std[:, 1], X_std[:, 2], 'o', markersize=8, color='red', alpha=0.5)
ax2.plot(lib[:, 0], lib[:, 1], lib[:, 2], '^', markersize=8, color='blue', alpha=0.5)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

X_pca2 = w[:, [0,2]].T.dot(X_std.T).T / np.sqrt(v[[0, 2]])
axes[0].scatter(X_pca2[:, 0], -1 * X_pca2[:, 1], edgecolors='black')
axes[0].set_title('my PCA')

pca2 = PCA(n_components=2, whiten=True)
lib2 = pca2.fit_transform(X_std)
axes[1].scatter(lib2[:, 0], lib2[:, 1], edgecolors='black')
axes[1].set_title('Sklearn PCA')
```
